```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from datetime import datetime
```

```python
 #2. تحميل البيانات
df = pd.read_csv('/content/OnlineRetail.csv')
```

```python
df = df[df['CustomerID'].notnull()]
df = df[df['Quantity'] > 0]
```

```python
# 4. إنشاء عمود الإيرادات (Total Price)
df['TotalPrice'] = df['Quantity'] * df['UnitPrice']
df.head()
```

```python
# 5. تحديد تاريخ التحليل ( آخر تاريخ بالبيانات + يوم)
snapshot_date = pd.to_datetime(df['InvoiceDate']).max() + pd.Timedelta(days=1)

# تحويل عمود التاريخ لتاريخ
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])
```

```python
# 6. حساب الـRFM لكل عميل
rfm = df.groupby('CustomerID').agg({
    'InvoiceDate': lambda x: (snapshot_date - x.max()).days,
    'InvoiceNo': 'count',
    'TotalPrice': 'sum'
})

rfm.rename(columns={'InvoiceDate': 'Recency',
                    'InvoiceNo': 'Frequency',
                    'TotalPrice': 'Monetary'}, inplace=True)
```

```python
# 7. حساب Scores لكل عنصر (1–5)
rfm['R_Score'] = pd.qcut(rfm['Recency'], 5, labels=[5,4,3,2,1])
rfm['F_Score'] = pd.qcut(rfm['Frequency'].rank(method='first'), 5, labels=[1,2,3,4,5])
rfm['M_Score'] = pd.qcut(rfm['Monetary'], 5, labels=[1,2,3,4,5])
```

```python
# 8. عمل RFM_Segment
rfm['RFM_Segment'] = rfm['R_Score'].astype(str) + rfm['F_Score'].astype(str) + rfm['M_Score'].astype(str)
```

```python
 #9. حساب متوسط RFM Score
rfm['RFM_Score'] = rfm[['R_Score','F_Score','M_Score']].astype(int).sum(axis=1)
```

```python
# 10. تقسيم العملاء بناءً على RFM_Score
def segment_customer(score):
    if score >= 12:
        return 'Champions'       # عملاء مميزين
    elif score >= 9:
        return 'Loyal Customers' # عملاء أوفياء
    elif score >= 6:
        return 'Potential'       # ممكن يكونوا أوفياء
    else:
        return 'At Risk'         # معرضين للفقدان

rfm['Segment'] = rfm['RFM_Score'].apply(segment_customer)
```

```python
seg_summary = rfm.groupby('Segment').agg({
    'Recency':'mean',
    'Frequency':'mean',
    'Monetary':'mean'
})

seg_summary['CustomerCount'] = rfm.groupby('Segment').size()

print(seg_summary)
```

```python
# 12. Visualization - عدد العملاء في كل Segment
plt.figure(figsize=(8,5))
sns.countplot(x='Segment', data=rfm, order=rfm['Segment'].value_counts().index)
plt.title('Customer Segments')
plt.show()
```

```python
# 13. Visualization Heatmap
rfm_heatmap = rfm.pivot_table(index='F_Score', columns='R_Score', values='Monetary', aggfunc='mean')
plt.figure(figsize=(8,6))
sns.heatmap(rfm_heatmap, annot=True, fmt=".0f", cmap='Blues')
plt.title('Average Monetary by R and F Scores')
plt.show()
```
